```python
"""
DOT Product: Associative and Distributive properties 

# FORMULAS:
let take a and b to be vectors 

therefore: a.b = a1b1 + a2b2 + a3b3 + a4b4
also:      a.b = |a||b|cosθ

# Dot product is distributive and not associative 
# in algebra, distributive means: a(b + c) = ab + ac where a, b, and c are normal scalars
    # You basically distribute a amongst b and c

# In linear algebbra:
    # lets take vector A and take its transpose = aT
    # aT(b+c) = aTb + aTc where b and c are also vectors 
# Dot product is not associative; i.e a(b.c) 
    #aT(bTc) != (aTb)Tc # you cannot just put brackets anywhere

# Matrix multliplication is associative: A(BC) = C(BA)
"""
```

```python
print("DEALING WITH DOT/SCALAR PRODUCT")  
print('')
print("Scalar as an answer")
print('')
print("let take a and b to be vectors")
print('')
print("a.b = a1b1 + a2b2 + a3b3 + a4b4")
print('')
print("a.b = |a||b|cosθ")
print('')
# Distributive property
# Generating vectors randomly
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)
"""
print(a)
print(b)
print(c)
"""
# computing dot product: a(b+c) == ab + ac
res1 = np.dot(a, (b + c))
res2 = np.dot(a,b) + np.dot(a,c) 

# comparing the two
print([res1, res2]) # equal, distributive property confirmed 
```

```python
"""
Is the Dot product commutative?
aT is a transpose and bT is b transpose 
dot product = Transpose of the first vector times the second vector and is always a single number
"""
print("Is the dot product commutative? i.e aT*b == bT*a")

# generating a 100 elements row vector
a = np.random.randn(100)
b = np.random.randn(100)

# computing aTb and bTa
dp_ab = np.dot(a,b)
dp_ba = np.dot(b,a)
print(dp_ab, dp_ba, dp_ab-dp_ba) 
print('')

v = [2, 4]
w = [3, 5]
print(np.dot(v,w), np.dot(w, v), np.dot(v,w) - np.dot(w, v))
```

```python
print(f"\u0332".join("THE OUTER PRODUCT"))
print("""
Unlike the dot product, outer product is not a single number
but a matrix. For dot product, you transpose the first vector while 
for outer product you transpose the second vector. 
For the Examples below, T denotes transpose

Let's say we have two vectors: A and B

Dot Product: (A.T)B = Transpose the first vector then multiply by the second
to get a scalar answer 

Outer Product: A(B.T) = Transpose the second vector then multiply it by the first
vector to get a matrix answer
""")

A = np.array([1, 2, 3, 4])
B = np.array([1, 2, 3, 4])
print(f"The outer product of vector A: {A} \nand vector B: {B} \nis given by np.outer(A,B) \nand the answer is:\n{np.outer(A,B)}\n\n")

#'Method 2'
def computeOuterProduct(vec1,vec2):
    outerProduct = np.zeros((len(A),len(B)))
    for i in range(0,len(A)):
        for j in range(0,len(B)):
            outerProduct[i,j] = A[i]*B[j]
    return outerProduct
print(computeOuterProduct(A,B))


```

```python
print(f"\u0332".join("VECTOR CROSS PRODUCT"))
print("""
This is the third way to multiply vectors besides dot product and outer product
Only defined for two vectors each containing 3 elements

Does not exist for 2 elements or 4 elements vectors. Only 3 element vectors

    1           a                            2c-3b
A = 2    and B =b  then the cross_product C =3a-1c
    3           c                            1b-2a
    
So HERE IS THE CATCH: The cross product is orthogonal to vector A and B and pops out of the plane
and is in right angle with the entire plane. It is sometimes called the normal to the plane
or the normal vector of that plane.
""")

# Creating vectors A and B
A = [-3, 2, 5]
B = [4, -3, 0]

# Using python cross product function 
crossProduct = np.cross(A,B)
print(f"The cross product computed using inbuilt python function crossProdut = np.cross(A,B) is:\n\n{crossProduct}")

# Using Manual Method
def crossProductManualMethod(vec1,vec2):
    sol = [ A[1]*B[2]-A[2]*B[1],
          A[2]*B[0]-A[0]*B[2],
          A[0]*B[1]-A[1]*B[0]]
    return sol

print(crossProductManualMethod(A,B))

# Plotting
fig = plt.figure()
ax = fig.gca(projection='3d')

# draw plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10,10,10), np.linspace(-10,10,10))
z1 = (-crossProduct[0]*xx-crossProduct[1]*yy)/crossProduct[2]
ax.plot_surface(xx,yy,z1)

## Plot the two vectors
ax.plot([0, A[0]], [0, A[1]], [0, A[2]], 'k')
ax.plot([0, B[0]], [0, B[1]], [0, B[2]], 'k')
ax.plot([0, crossProduct[0]], [0, crossProduct[1]], [0, crossProduct[2]], 'r')

ax.view_init(azim=150, elev=45)
plt.show()
```

```python
print(f"\u0332".join("VECTORs WITH COMPLEX NUMBERS"))
print("""
Important point: Complex numbers and the imaginary operator 
are very useful for applications in signal processing,
engineering, applied math, etc

Use real and imaginary axis coordinates (2 3i) refers to a 
single complex number with 2 in real part and 3 in imaginary 
part. They are useful because they contain a lot of information.
They have a magnitude away from the complex plane and an angle from 
the real axis. Its like a 2d vector all on its own

z = a + bi
w = c + di
zw = (a + bi)(c + di)
   = ac + adi + cbi + bdi^2
   = ac + adi + cbi - bd
   
""")

print(f"\u0332".join("HERMITIAN TRANSPOSE AKA CONJUGATE TRANSPOSE AKA COMPLEX CONJUGATE TRANSPOSE"))
print("""

It simply entails flipping the sign of the conjugate part of the complex number
Indicated using H or * 
Z = 2 + 3i
Z*= 2 - 3i

Geometric interpretation is reflection across the real axis. 

WHY DO WE NEED HERMITIAN TRANSPOSE???
(Z*Z)^0.5 or (Z^H*.Z)^0.5 gives us the magnitude of the vector which is a real number

""")
print(f"\u0332".join("HERMITIAN TRANSPOSE AKA CONJUGATE TRANSPOSE AKA COMPLEX CONJUGATE TRANSPOSE COMPUTATIONS"))

# Creating complex number:
z = np.complex(3,4)
print(f"Complex number: z = {z}")

# Magnitude 
zMagnitude = np.linalg.norm(z)
print(f"The magnitude of z is: {zMagnitude}")
print('')
print(f"Magnitude by transpose: {np.transpose(z)*z}")
print('')
print(f"Magnitude by Hermitian Transpose: {np.transpose(z.conjugate())*z}")
print('')

# Complex Vector
complexVector = np.array([3,4j,5+2j, np.complex(2,-5)])
print(f"The complex vector is: {complexVector}")
print('')
print(f"The transpose of the above complex vector is: {np.transpose(complexVector)}")
print('')
print(f"The hermitian transpose of the complex vector is: {np.transpose(complexVector.conjugate())}")
print('')
print(f"The magnitude of the complex vector is: {complexVector*(np.transpose(complexVector.conjugate()))}")

```

```python
print(f"\u0332".join("INTERPRETING AND CREATING UNIT VECTORS\n"))
print("""
Unit Vector - Has a length of 1

μv such that ||μv|| = 1

But how to choose μ? - you multiply any number by its reciprocal
μ3 = 1 ~ to turn 3 into a unit vector; you multiply by μ whose value is the reciprocal of 3
Therefore
μ = 1/3
μ = 1/||v|| i.e i/[magnitude of vector v]

COMPUTATIONS:\n
""")

# vector 
vector = np.array([-3, 6])
print(f"Vector is: {vector}\n")
print(f"μ is given by 1/np.linalg.norm(vector) which is : {1/(np.linalg.norm(vector))}\n")
print(f"Converting vector to unit vector is given by vector x mu which gives: {vector*(1/(np.linalg.norm(vector)))}\n")
print(f"The magnitude of the unit vector is now: {np.linalg.norm(vector*(1/(np.linalg.norm(vector))))}")

# visualization
print(f"\u0332".join("VISUALIZATION OF UNIT VECTOR IN 2 DIMENSIONAL PLANE\n"))
# Creating variables of the above statements for simpicity
μ = 1/(np.linalg.norm(vector))
unitVector = vector*μ

plt.plot([0,unitVector[0]], [0,unitVector[1]], 'r', label='vector-norm', linewidth=5)
plt.plot([0, vector[0]], [0, vector[1]], 'b', label='vector')
plt.axis('square')
plt.axis((-8,8,-8,8)) # Don't forget two brackets
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()


```
