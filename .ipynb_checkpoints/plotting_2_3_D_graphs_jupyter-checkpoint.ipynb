```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#2 dimensional vector 
v2 = np.array([3, -2])

#3 dimensional vector 
v3 = np.array([4, -3, 2])

#TRamsforming row vector to column vector & vice-versa
```

```python
#PLotting 
plt.plot([0,v2[0]], [0,v2[1]])
plt.axis('equal')
plt.plot([-4, 4], [0,0], 'k--')
plt.plot([0, 0], [-4, 4], 'k--')
plt.grid()
plt.axis((-4, 4, -4, 4))
plt.show()
```

```python
#3D Plotting in python 
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(projection='3d')
ax.plot([0, v3[0]],[0, v3[1]], [0, v3[2]], linewidth=3)

#Beautifying the plot 
ax.plot([0,0], [0,0], [-4,4], 'k--')
ax.plot([0,0], [-4,4], [0,0], 'k--')
ax.plot([-4,4], [0,0], [0,0], 'k--')
plt.show()
```

```python
"""
DOT Product: Associative and Distributive properties 

# FORMULAS:
let take a and b to be vectors 

therefore: a.b = a1b1 + a2b2 + a3b3 + a4b4
also:      a.b = |a||b|cosθ

# Dot product is distributive and not associative 
# in algebra, distributive means: a(b + c) = ab + ac where a, b, and c are normal scalars
    # You basically distribute a amongst b and c

# In linear algebbra:
    # lets take vector A and take its transpose = aT
    # aT(b+c) = aTb + aTc where b and c are also vectors 
# Dot product is not associative; i.e a(b.c) 
    #aT(bTc) != (aTb)Tc # you cannot just put brackets anywhere

# Matrix multliplication is associative: A(BC) = C(BA)
"""
```

```python
print("DEALING WITH DOT/SCALAR PRODUCT")  
print('')
print("Scalar as an answer")
print('')
print("let take a and b to be vectors")
print('')
print("a.b = a1b1 + a2b2 + a3b3 + a4b4")
print('')
print("a.b = |a||b|cosθ")
print('')
# Distributive property
# Generating vectors randomly
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)
"""
print(a)
print(b)
print(c)
"""
# computing dot product: a(b+c) == ab + ac
res1 = np.dot(a, (b + c))
res2 = np.dot(a,b) + np.dot(a,c) 

# comparing the two
print([res1, res2]) # equal, distributive property confirmed 
```

```python
# Dot products with matrix columns 
m = 2
n = 3
A = np.random.randn(m,n) # 4x6 matrices of random numbers
B = np.random.randn(m,n)
print(A)
print('')
print(B)

dps = np.zeros(3) # creating an empty 6 element vector. Remember, each column is a vector
for i in range(3):
    dps[i] = np.dot( A[:,i], B[:,i] )
print('')
print(dps)
```

```python
"""
Is the Dot product commutative?
aT is a transpose and bT is b transpose 
"""
print("Is the dot product commutative? i.e aT*b == bT+a")

# generating a 100 elements row vector
a = np.random.randn(100)
b = np.random.randn(100)

# computing aTb and bTa
dp_ab = np.dot(a,b)
dp_ba = np.dot(b,a)
print(dp_ab, dp_ba, dp_ab-dp_ba) 
print('')

v = [2, 4]
w = [3, 5]
print(np.dot(v,w), np.dot(w, v), np.dot(v,w) - np.dot(w, v))
```
