```python
print(f"\u0332".join("THE OUTER PRODUCT"))
print("""
Unlike the dot product, outer product is not a single number
but a matrix. For dot product, you transpose the first vector while 
for outer product you transpose the second vector. 
For the Examples below, T denotes transpose

Let's say we have two vectors: A and B

Dot Product: (A.T)B = Transpose the first vector then multiply by the second
to get a scalar answer 

Outer Product: A(B.T) = Transpose the second vector then multiply it by the first
vector to get a matrix answer
""")

A = np.array([1, 2, 3, 4])
print(A)
B = np.array([1, 2, 3, 4])
print(f"The outer product of vector A: {A} \nand vector B: {B} \nis given by np.outer(A,B) \nand the answer is:\n{np.outer(A,B)}\n\n")

#'Method 2'
def computeOuterProduct(vec1,vec2):
    outerProduct = np.zeros((len(A),len(B)))
    for i in range(0,len(A)):
        for j in range(0,len(B)):
            outerProduct[i,j] = A[i]*B[j]
    return outerProduct
print(computeOuterProduct(A,B))


```

```python
print(f"\u0332".join("VECTOR CROSS PRODUCT"))
print("""
This is the third way to multiply vectors besides dot product and outer product
Only defined for two vectors each containing 3 elements

Does not exist for 2 elements or 4 elements vectors. Only 3 element vectors

    1           a                            2c-3b
A = 2    and B =b  then the cross_product C =3a-1c
    3           c                            1b-2a
    
So HERE IS THE CATCH: The cross product is orthogonal to vector A and B and pops out of the plane
and is in right angle with the entire plane. It is sometimes called the normal to the plane
or the normal vector of that plane.
""")

# Creating vectors A and B
A = [-3, 2, 5]
B = [4, -3, 0]

# Using python cross product function 
crossProduct = np.cross(A,B)
print(f"The cross product computed using inbuilt python function crossProdut = np.cross(A,B) is:\n\n{crossProduct}")

# Using Manual Method
def crossProductManualMethod(vec1,vec2):
    sol = [ A[1]*B[2]-A[2]*B[1],
          A[2]*B[0]-A[0]*B[2],
          A[0]*B[1]-A[1]*B[0]]
    return sol

print(crossProductManualMethod(A,B))

# Plotting
fig = plt.figure()
ax = fig.gca(projection='3d')

# draw plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10,10,10), np.linspace(-10,10,10))
z1 = (-crossProduct[0]*xx-crossProduct[1]*yy)/crossProduct[2]
ax.plot_surface(xx,yy,z1)

## Plot the two vectors
ax.plot([0, A[0]], [0, A[1]], [0, A[2]], 'k')
ax.plot([0, B[0]], [0, B[1]], [0, B[2]], 'k')
ax.plot([0, crossProduct[0]], [0, crossProduct[1]], [0, crossProduct[2]], 'r')

ax.view_init(azim=150, elev=45)
plt.show()
```

```python
print(f"\u0332".join("INTERPRETING AND CREATING UNIT VECTORS\n"))
print("""
Unit Vector - Has a length of 1

μv such that ||μv|| = 1

But how to choose μ? - you multiply any number by its reciprocal
μ3 = 1 ~ to turn 3 into a unit vector; you multiply by μ whose value is the reciprocal of 3
Therefore
μ = 1/3
μ = 1/||v|| i.e i/[magnitude of vector v]

COMPUTATIONS:\n
""")

# vector 
vector = np.array([-3, 6])
print(f"Vector is: {vector}\n")
print(f"μ is given by 1/np.linalg.norm(vector) which is : {1/(np.linalg.norm(vector))}\n")
print(f"Converting vector to unit vector is given by vector x mu which gives: {vector*(1/(np.linalg.norm(vector)))}\n")
print(f"The magnitude of the unit vector is now: {np.linalg.norm(vector*(1/(np.linalg.norm(vector))))}")

# visualization
print(f"\u0332".join("VISUALIZATION OF UNIT VECTOR IN 2 DIMENSIONAL PLANE\n"))
# Creating variables of the above statements for simpicity
μ = 1/(np.linalg.norm(vector))
unitVector = vector*μ

plt.plot([0,unitVector[0]], [0,unitVector[1]], 'r', label='vector-norm', linewidth=5)
plt.plot([0, vector[0]], [0, vector[1]], 'b', label='vector')
plt.axis('square')
plt.axis((-8,8,-8,8)) # Don't forget two brackets
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()


```

```python
print(f"\u0332".join("SPAN\n"))
print("""
Watch Vid 28 of linear algebra
A span of set of vectors: i.e span({v1,....,vn}) = α1v1+.....αnvn, α∈R
                        : is all possible linear combinations of all the vectors in that set
                        
Span stretches to infinity
""")

# Defining two vectors with the mind that they are in a set
s1 = np.array([1, 1, 0])
s2 = np.array([1, 7, 0])

# Defining vectors v and w to answer: what is the geometric interpretation of vector v being in a set spanned by s1 and s2?
v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

# Drawing vectors
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0,s1[0]], [0, s1[1]], [0, s1[2]+0.1], 'r', linewidth=3)
ax.plot([0,s2[0]], [0, s2[1]], [0, s2[2]+0.1], 'r', linewidth=3)

ax.plot([0, v[0]], [0, v[1]], [0, v[2]+0.1], 'g', linewidth=3)
ax.plot([0, w[0]], [0, w[1]], [0, w[2]], 'b')

# Drawing line
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(s1,s2)
z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2]
ax.plot_surface(xx,yy,z1)
plt.show()
```
