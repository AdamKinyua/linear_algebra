```python
import numpy as np
```

```python
# Square and Rectangular Matrices
squareMatrix = np.random.randn(5,5)
rectangleMatrix = np.random.randn(5,2)
print(f"SQUARE MATRIX \n\n{squareMatrix}\n\n RECTANGULAR MATRIX\n\n{rectangleMatrix}\n\n")

# Creating Identity matrix
identityMatrix = np.eye(3)
print(f"IDENTITY MATRIX\n\n {identityMatrix}\n\n")

# Creating Zeros matrix
zerosMatrix = np.zeros((4,4))
print(f"ZEROS MATRIX\n\n {zerosMatrix}\n\n")

# Creating Zeros matrix with complex data type
zerosMatrix = np.zeros((4,4), dtype=complex)
print(f"COMPLEX ZEROS MATRIX\n\n {zerosMatrix}\n\n")

# Diagonal Matrix 
diagonalMatrix = np.diag([1,2,3,4,2,1])
print(f"DIAGONAL MATRIX\n\n {diagonalMatrix}\n\n")

# Creating triangular matrix from full matrix 
matFile = np.random.randn(5,5,)
upperTriangularMatrix = np.triu(matFile)
lowerTriangualarMatrix = np.tril(matFile)
print(f"From matrix\n {matFile}\n\n We created upper triangular matrix\n\n {upperTriangularMatrix}\n\n and lower triangular matrix\n\n {lowerTriangualarMatrix}\n\n")

# Concatenate matrices (sizes must match in terms of rows)
matrix1 = np.random.randn(3,2)
matrix2 = np.random.randn(3,4)
concatenatedMatrix = np.concatenate((matrix1,matrix2),axis=1)
print(f"Concatenated matrix of\n\n {matrix1} \n\n and \n\n {matrix2} \n\n is \n\n{concatenatedMatrix}\n\n")

# Matrix transpose - switch rows and columns 
# Symmetric (A=A^T) and skew symmetric (A=-A^T)
symm = np.array([[1, -1, 0],[-1, -2, -4],[0,-4,5]])
skewSymm = np.array([[0, 1, -2],[-1,0,-4],[2,4,0]])
print(f"SYMMETRIC MATRIX\n\n{symm}\n\n SKEW-SYMMETRIC \n\n {skewSymm}")


```
